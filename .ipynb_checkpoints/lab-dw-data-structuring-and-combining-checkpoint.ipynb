## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd

#1. Load data
df_insurance_file2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df_insurance_file3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')

#2. Data overview
#print('File 2 overview')
#print(df_insurance_file2.head())

#print('File 3 overview')
#print(df_insurance_file3.head())


#3. Cleaning data

#3.1.Column names should be in lower case
df_insurance_file2.columns = df_insurance_file2.columns.str.lower()
df_insurance_file3.columns = df_insurance_file3.columns.str.lower()

#3.2. White spaces in column names should be replaced by _
df_insurance_file2.columns = df_insurance_file2.columns.str.replace(' ', '_')
df_insurance_file3.columns = df_insurance_file3.columns.str.replace(' ', '_')

#3.3. ST should be replaced for state
df_insurance_file2.rename(columns={'st': 'state'}, inplace=True)

#3.4. Gender standardization
gender_replacements = {
    'F': 'Female',
    'M': 'Male',
    'Femal': 'Female',
    'female': 'Female',
    'Male': 'Male'
}
df_insurance_file2['gender'] = df_insurance_file2['gender'].replace(gender_replacements)
df_insurance_file3['gender'] = df_insurance_file3['gender'].replace(gender_replacements)

print("\nValues after standardization of the column Gender:")
print(df_insurance_file2['gender'].unique())
print(df_insurance_file3['gender'].unique())


#3.5. State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
  #Gender standardization
state_replacements = {
    'AZ': 'Arizona',
    'Cali': 'California',
    'WA': 'Washington'
}

df_insurance_file2['state'] = df_insurance_file2['state'].replace(state_replacements)
df_insurance_file3['state'] = df_insurance_file2['state'].replace(state_replacements)

print("\nValues after standardization of the column State:")
print(df_insurance_file2['state'].unique())
print(df_insurance_file2['state'].unique())


#3.6. In education, "Bachelors" could be replaced by "Bachelor"

df_insurance_file2['education'] = df_insurance_file2['education'].replace('Bachelors', 'Bachelor')
df_insurance_file3['education'] = df_insurance_file3['education'].replace('Bachelors', 'Bachelor')

print("\nValues after standardization of the column Education:")
print(df_insurance_file2['education'].unique())
print(df_insurance_file3['education'].unique())

#3.7. In Customer Lifetime Value, delete the % character
df_insurance_file2['customer_lifetime_value'] = df_insurance_file2['customer_lifetime_value'].astype(str)
df_insurance_file3['customer_lifetime_value'] = df_insurance_file3['customer_lifetime_value'].astype(str)

print("Data type after the modification (from float to string):")
print(df_insurance_file2['customer_lifetime_value'].dtype)
print(df_insurance_file3['customer_lifetime_value'].dtype)

#3.8. Customer Lifetime Value remplacements and rechanging the data type (from string to float)
df_insurance_file2['customer_lifetime_value'] = df_insurance_file2['customer_lifetime_value'].str.replace('%', '').astype(float)
df_insurance_file3['customer_lifetime_value'] = df_insurance_file3['customer_lifetime_value'].str.replace('%', '').astype(float)

#3.9. In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"

  #Vehicle classs standardization
vehicle_class_replacements = {
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
}

df_insurance_file2['vehicle_class'] = df_insurance_file2['vehicle_class'].replace(vehicle_class_replacements)
df_insurance_file3['vehicle_class'] = df_insurance_file3['vehicle_class'].replace(vehicle_class_replacements)

  #Print new values in column vehicle class
print("\nValuesafter standardization of the column Vehicule class:")
print(df_insurance_file2['vehicle_class'].unique())
print(df_insurance_file3['vehicle_class'].unique())

#4.Changing the format of Number of open complaints:

#4.1.Replacing the value "nan" by 0/0/0
df_insurance_file2['number_of_open_complaints'] = df_insurance_file2['number_of_open_complaints'].fillna('0/0/0')
df_insurance_file3['number_of_open_complaints'] = df_insurance_file3['number_of_open_complaints'].fillna('0/0/0')

#4.2.chaging the data type from object to string
df_insurance_file2['number_of_open_complaints'] = df_insurance_file2['number_of_open_complaints'].astype(str)
df_insurance_file3['number_of_open_complaints'] = df_insurance_file3['number_of_open_complaints'].fillna('0/0/0')

#4.3.removing the slash: (préciser le fonctionnement)
df_insurance_file2['number_of_open_complaints'] = df_insurance_file2['number_of_open_complaints'].str.split('/').str[1]
df_insurance_file3['number_of_open_complaints'] = df_insurance_file3['number_of_open_complaints'].fillna('0/0/0')

#4.4.chaging the data type from string to integer
df_insurance_file2['number_of_open_complaints'] = df_insurance_file2['number_of_open_complaints'].fillna('0').astype(int)
df_insurance_file3['number_of_open_complaints'] = df_insurance_file3['number_of_open_complaints'].fillna('0/0/0')

#4.5.print the data type print the head after modifcations
print("\nData type for Number of Open Complaints after modificatons :")
print(df_insurance_file2['number_of_open_complaints'].dtype)
print(df_insurance_file3['number_of_open_complaints'].dtype)

#4.6.print the head after modifcations
print("Example of the values stored in the column Number of Open Complaints after modifications:")
print(df_insurance_file2['number_of_open_complaints'].head())
print(df_insurance_file3['number_of_open_complaints'].head())

#5.Identify the columns with null values 
print("Null values before filling :")
print(df_insurance_file2.isnull().sum())
print(df_insurance_file3.isnull().sum())

#5.1. Determine numerical and categorical variables (if data type is float or interger --> numerical, and if data type is object --> categorical)
num_cols = df_insurance_file2.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df_insurance_file3.select_dtypes(include=['object']).columns

#5.2. Fill the null values

  #Numerical columbs: filling with the mean
for col in num_cols:
    df_insurance_file2[col] = df_insurance_file2[col].fillna(df_insurance_file2[col].mean())
    df_insurance_file3[col] = df_insurance_file3[col].fillna(df_insurance_file3[col].mean())

  #Pour les colonnes catégorielles, remplir avec le mode
for col in cat_cols:
    df_insurance_file2[col] = df_insurance_file2[col].fillna(df_insurance_file2[col].mode()[0])
    
for col in cat_cols:
    df_insurance_file3[col] = df_insurance_file3[col].fillna(df_insurance_file3[col].mode()[0])

  #Verify that Null Values have been Correctly Handled
print("\nNull values after filling :")
print(df_insurance_file2.isnull().sum())
print(df_insurance_file3.isnull().sum())

```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd

#1. Load the data
df_marketing_customer = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')

#2. Data overview
print('Marketing customer file overview:')
print(df_marketing_customer.head())
print(df_marketing_customer.info())

#3. Checking missing values
missing_values = df_marketing_customer.isnull().sum()
print("\nMissing values per columns:")
print(missing_values)


# Examen des valeurs des colonnes numériques
numeric_columns = df_marketing_customer.select_dtypes(include=['int', 'float']).columns
print("\nExamen des valeurs des colonnes numériques :")
for col in numeric_columns:
    non_numeric_values = df_marketing_customer[col].apply(lambda x: isinstance(x, str))
    if non_numeric_values.any():
        print(f"Colonnes '{col}' contient des valeurs non numériques :")
        print(df_marketing_customer[col][non_numeric_values])

# Conversion des colonnes numériques
print("\nConversion des colonnes numériques :")
for col in numeric_columns:
    try:
        df_marketing_customer[col] = pd.to_numeric(df_marketing_customer[col])
        print(f"Colonnes '{col}' convertie avec succès.")
    except ValueError as e:
        print(f"Erreur lors de la conversion de la colonne '{col}' : {e}")

# Remplacement des valeurs incorrectes ou manquantes
# Dans cet exemple, nous pouvons remplacer les valeurs incorrectes par NaN
df_marketing_customer.rename(columns={"employmentstatus": "employment_status"}, inplace=True)
df_marketing_customer[numeric_columns] = df_marketing_customer[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Affichage du DataFrame avec les valeurs remplacées
print("\nDataFrame avec les valeurs remplacées :")
print(df_marketing_customer)

# Compter le nombre de valeurs NaN par colonne
nan_count_by_column = df_marketing_customer.isna().sum()

# Afficher le nombre de valeurs NaN par colonne
print("Nombre de valeurs NaN par colonne :")
print(nan_count_by_column)



```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Calculer le total des ventes par canal de vente
sales_channel_revenue = df_marketing_customer.pivot_table(index='sales_channel', values='total_claim_amount', aggfunc='sum').round(2)

# Afficher le tableau récapitulatif
print("Total revenue for each sales channel:")
print(sales_channel_revenue)

# Créer une table pivotante montrant la valeur moyenne de la durée de vie du client par genre et niveau d'éducation
avg_clv_pivot = df_marketing_customer.pivot_table(index=['gender', 'education'], values='customer_lifetime_value', aggfunc='mean').round(2)

# Afficher la table pivotante
print("\nAverage customer lifetime value per gender and education level:")
print(avg_clv_pivot)
```
